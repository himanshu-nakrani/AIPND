# Absolute vs. Relative Frequency

By default, seaborn's countplot function will summarize and plot the data in terms of absolute frequency, or pure counts. In certain cases, you might want to understand the distribution of data or want to compare levels in terms of the proportions of the whole. In this case, you will want to plot the data in terms of relative frequency, where the height indicates the proportion of data taking each level, rather than the absolute count.



One method of plotting the data in terms of relative frequency on a bar chart is to just relabel the count's axis in terms of proportions. The underlying data will be the same, it will simply be the scale of the axis ticks that will be changed.



```python
pkmn_types = data.melt(id_vars=['id', 'species'], 
                          value_vars=['type_1', 'type_2'], 
                          var_name='type_level', 
                          value_name='type')
pkmn_types.head(10)
```

***pandas.DataFrame.melt()*** - method to unpivot a DataFrame from wide to long format, optionally leaving identifiers set. 

**Syntax** - DataFrame.melt(id_vars, value_vars, var_name, value_name, col_level, ignore_index)

1. *id_vars* - It is a tuple representing the column(s) to use as identifier variables.
2. *value_vars *- It is tuple representing the column(s) to unpivot (remove, out of place).
3. *var_name* - It is a name of the new column.
4. *value_name* - It is a name to use for the ‘value’ of the columns that are unpivoted.

```python
# Counting the frequency of unique values in the `type` column of pkmn_types dataframe. 
# By default, returns the decreasing order of the frequency.
```

```python
type_counts = pkmn_types['type'].value_counts()
type_counts
```

```python
# Getting the unique values of the `type` column, in the decreasing order of the frequency.
```

```python
type_order = type_counts.index
type_order
```

```python
#  Plotting the horizontal bar charts
```

```python
base_color = sns.color_palette()[0]
sns.countplot(data=pkmn_types, y='type', color=base_color, order=type_order);
```

```python
# Plotting a bar chart having the proportions, instead of the actual count, on one of the axes.
```

```python
n_pokemon = pkmn_types['type'].value_counts().sum()

max_type_count = type_counts[0]

max_prop = max_type_count / n_pokemon
print(max_prop)
```

```python
# Creating an array of evenly spaced proportioned values
```

```python
tick_props = np.arange(0, max_prop, 0.02)
tick_props
```

```python
# Creating a list of String values that can be used as tick labels.
```

```python
#  comprehension to create tick_names that we will apply to the tick labels.
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
tick_names
```

```python
# Plotting the bar chart, with new x-tick labels
```

```python
sns.countplot(data=pkmn_types, y='type', color=base_color, order=type_order);

# Change the tick locations and labels
plt.xticks(tick_props * n_pokemon, tick_names)
plt.xlabel('proportion');
```

```python
# Printing the proportion on the bars of a horizontal plot.
```

```python
# Considering the same chart from the Example 1 above, print the text (proportion) on the bars
base_color = sns.color_palette()[0]
plt.figure(figsize = (20, 12))
sns.countplot(data=pkmn_types, y='type', color=base_color, order=type_order);

for i in range (type_counts.shape[0]):
    count = type_counts[i]
    pct_string = '{:0.1f}'.format(100*count/n_pokemon)
    plt.text(count+1, i, pct_string, va='center')
```

# Pie Charts

common univariate plot type that is used to depict relative frequencies for levels of a categorical variable.

Frequencies in a pie chart are depicted as wedges drawn on a circle: the larger the angle or area, the more common the categorical value taken. 

**Use only when -** 
 1. when the number of categories is less
 2. you'd like to see the proportion of each category on a chart
 
 

**Note -**
- Make sure that your interest is in relative frequencies. 
- Limit the number of slices plotted. A pie chart looks best with two or three slices. If you have a lot of categories, or categories then you can group categories that has small proportional representation and plot them as 'Other' category.
- Plot the data systematically. One typical method of plotting a pie chart is to start from the top of the circle, then plot each categorical level clockwise from most frequent to least frequent. If you have three categories and are interested in the comparison of two of them, a common plotting method is to place the two categories of interest on either side of the 12 o'clock direction, with the third category filling in the remaining space at the bottom.

**matplotlib.pyplot.pie(x_data, labels, colors, startangle, counterclock, wedgeprops)**

**matplotlib.pyplot.axis()** - function to set some axis properties.

### Plot a simple Pie chart

```python
sorted_counts = data['generation_id'].value_counts()

plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False);

plt.axis('square')
```

```python
sorted_counts = data['generation_id'].value_counts()

plt.pie(sorted_counts, labels = ['gen5', 'gen1', 'gen3', 'gen4', 'gen2', 'gen7', 'gen6'], startangle = 90, counterclock = False);

plt.axis('square')
```

## Donut Plot

 It's just like a pie chart, except that there's a hole in the center of the plot.

```python
sorted_counts = data['generation_id'].value_counts()

plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False, wedgeprops = {'width' : 0.4});
plt.axis('square')
```

# Figures, Axes, and Subplots

The base of visualization in matplotlib is a Figure object. Contained within each Figure will be one or more Axes objects, each Axes object containing a number of other elements that represent each plot. In the earliest examples, these objects have been created implicitly. 

Suppose we are creating a histogram, then following steps will be performed by Python:

Python first creates a Figure object. And since the Figure doesn't start with any Axes to draw the histogram onto, an Axes object is created inside the Figure. 

```python
#  explicitly set up the Figure and Axes like this
```

```python
# matplotlib
```

```python
fig = plt.figure()    # Creating a new figure

# The argument represents the dimensions [left, bottom, width, height] 
ax = fig.add_axes([.125, .125, .775, .755])
ax.hist(data=data, x='speed');
```

# Scales and Transformations

```python
# Scale the x-axis to log-type
```

```python
plt.figure(figsize = [20, 5])

plt.subplot(1, 2, 1)
plt.hist(data=data, x='weight');
plt.xlabel('Initial plot with original data')


plt.subplot(1, 2, 2)
bins = np.arange(0, data['weight'].max()+40, 40)
plt.hist(data=data, x='weight', bins=bins);

plt.xscale('log')
plt.xlabel('The x-axis limits only scaled to log-type')
```

```python
# Scale the x-axis to log-type, and change the axis limit.
```

```python
bins = 10 ** np.arange(-1, 3+0.1, 0.1)
plt.hist(data=data, x='weight', bins=bins);

plt.xscale('log')

plt.xlabel('x-axis limits are changed, and scaled to log-type')
```

```python
# Scale the x-axis to log-type, change the axis limits, and increase the x-ticks
```

```python
bins = 10 ** np.arange(-1, 3+0.1, 0.1)

ticks = [0.1, 0.3, 1, 3, 10, 30, 100, 300, 1000]
labels = ['{}'.format(v) for v in ticks]

plt.hist(data=data, x='weight', bins=bins);

plt.xscale('log')

plt.xticks(ticks, labels);
```
